```python
import numpy as np
from fastRCNN_model import fast_RCNN_model
```

```python
img_dir = "../img/"
subclip_dir = [str(i*10)+"-"+str((i+1)*10)+"/" for i in range(23)]
video_num = [str(i) + " video " for i in [0, 1, 4, 5]]
frame_num = [str(i*30)+".png" for i in range(9)]
```

```python
model = fast_RCNN_model()
file = img_dir + subclip_dir[0] + video_num[0] + frame_num[0]
file2 = "../img/0-10/0 video 210.png"
file3 = "../img/0-10/1 video 210.png"
```

```python
start = 0
next_start = 0
result = []
```

```python
def edit():
    start = 0
    next_start = 0
    for period in subclip_dir[:10]:
        video_point = []
        start = next_start
        
        for f_num in frame_num:
            per_frame = []
            
            for v_num in video_num:
                file_name = img_dir + period + v_num + f_num
                print(file_name)
                model.write_file_name(file_name=file_name)
                box_id, box_score, box_val = model.run_model()
                point = np.sum(box_val) / 6000
                per_frame.append(point.asnumpy()[0])
            video_point.append(per_frame)
        
        frame_similarity = []
        selected_video = 0
        selected_time = 0
        
        for i in range(len(video_point)):
            temp = []
            for j in range(len(video_point[i])):
                if j != start:
                    temp.append(abs(video_point[i][start] - video_point[i][j]))
                else:
                    temp.append(100)
            frame_similarity.append(temp)
        
        edit_point = frame_similarity.index(min(frame_similarity))
        next_video = frame_similarity[edit_point].index(min(frame_similarity[edit_point]))
        next_start = next_video
        print("edit point: " + str(edit_point))
        print("next_video: " + str(next_start))
        print("____")
        r = [edit_point, next_video]
        result.append(r)
```
